```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

#data = np.loadtxt('stockholm_tmp.dat', dtype={'names':('year', 'month', 'day', 'mean', 'min', 'max'), 'formats':('i4', 'i4', 'i4', 'f4', 'f4', 'f4')})
data = np.loadtxt('stockholm_tmp.dat', dtype=float)
april = data[data[:,1] == 4]
april_mean = np.array([np.mean(april[april[:,0] == 1800+y][:,3]) for y in range(212)])
april_max = np.array([np.max(april[april[:,0] == 1800+y][:,3]) for y in range(212)])
april_min = np.array([np.min(april[april[:,0] == 1800+y][:,3]) for y in range(212)])
year = np.arange(1800,2012)
plt.xlabel('year')
plt.ylabel('temperature')
plt.title('Temperature in April')
plt.plot(year, april_mean[year-1800], 'g')
plt.plot(year, april_max[year-1800], 'r')
plt.plot(year, april_min[year-1800], 'b')
plt.legend(['mean', 'max', 'min'])
```

```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

ax = axes3d.Axes3D(plt.figure())

x = y = np.linspace(-3,3,74)
X,Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(4*R)/R
ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=cm.jet)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('anime.csv')
data = data.replace('?', np.NaN)
#print(data[:][0:10])
#types = data.dtypes
data.columns = data.columns.str.lower()
cstats = ['episodes', 'rating', 'voters']
#stats = pd.DataFrame({'column':cstats, 'min':[data[c].dropna().min() for c in cstats],'max':[data[c].dropna().max() for c in cstats]})

data.groupby('production')['title'].nunique()
#data.groupby('theme')['title'].nunique()
data.groupby('theme')['title'].nunique()
themes = ['Military', 'Cars', 'Police', 'Crunchyroll', 'Demons', 'Harem', 'School', 'Historical']
#[t in themes for theme in data['theme'].str.split(',') for t in theme]
data['theme'] = data['theme'].str.split(',')
#a = [[t in th for th in data['theme']] for t in themes]
sizes = [len(data.loc[data['theme'].apply(lambda l: t in l)]) for t in themes]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=themes, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
